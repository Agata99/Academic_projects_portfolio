Prognozujemy liczbę potwierdzonych przypadków koronawirusa w Polsce w dniach 19 – 25 kwietnia
2021 na podstawie danych z dwóch poprzednich tygodni. Korzystamy ze zbioru danych ze strony: https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv <br>
Stosujemy metodę średniej ruchomej ważonej, w której prognoza wartości zmiennej losowej $Y$ wyznaczana na okres $t$ jest średnią arytmetyczną ważoną wartości z poprzednich $k$ okresów. Model średniej ruchomej ważonej wyraża się wzorem
$$y_{t}^*=\displaystyle\sum_{i=t-k}^{t-1}y_{i}w_{i-t+k+1}.$$
Wagi dobieramy tak, aby spełniały poniższe założenia:
<ul>
<li>$w_i\geq 0,$</li>
<li>$w_i\leq w_{i+1},$</li>
<li>$\displaystyle\sum_{i=1}^{k}w_i=1.$</li>
</ul>
Obserwując dane, zauważamy, że dynamika liczby potwierdzonych przypadków zależna jest od części tygodnia. Dane napływające w weekendy lub w początkowej części tygodnia są z reguły niższe. Z tego powodu modyfikujemy metodę średniej ruchomej ważonej i prognozę wykonujemy w trzech częściach. Przewidywań dla poniedziałku i wtorku dokonujemy na podstawie liczby przypadków z niedziel, poniedziałków i wtorków z poprzednich dwóch tygodni. Środową i czwartkową wartość prognozujemy, biorąc pod uwagę znane dane z dwóch poprzednich sobót, środy, czwartku i piątku oraz ostatniej niedzieli. Uwzględniając znane wartości z piątku i weekendów oraz wcześniejsze prognozy, wyznaczamy prognozę dla piątku, soboty i niedzieli. Dobrane przez nas wagi to: $w_1=0.05, w_2=0.05, w_3=0.1, w_4=0.1, w_5=0.3, w_6=0.4$. W ostatnim kroku, w związku z zależnością liczby zachorowań od dnia tygodnia, prognozę na poniedziałek obniżamy o 20% i dodajemy 20% do wartości ze środy, czwartku, piątku i soboty.

```python
#Pobieramy listę łącznych zachorowań ze strony
import csv, urllib.request, datetime

url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
response = urllib.request.urlopen(url)
lines = [l.decode('utf-8') for l in response.readlines()]
cr = csv.reader(lines)

i=0
for row in cr:
    i=i+1
    if i==1:
        daty=row #Wybieramy wiersz z oznaczonymi datami
    if i==217:
        polska=row #Wybieramy wiersz z danymi Polski


for i in range(len(polska)-1,4,-1):
    polska[i]=int(polska[i])-int(polska[i-1]) #Przekształcamy dane łączne na dzienne

for i in range(0, len(polska)):
    print(daty[i],polska[i]) #Wyświetlamy pełną listę pobranych danych Polski
```

```python
#Podajemy ostatni dzień okresu, z którego pobierane są dane (data poprzedzająca pierwszy dzień prognozy)
dzien='18'
miesiac='4'
rok='2021'

#Podajemy długość pobranego okresu oraz ilość prognozowanych dni
ile_dni=14
ile_dni_prognozy=7

#Zamieniamy pobrane wartości na odpowiednie formatowanie
if len(rok)==4:
    rok=rok[2]+rok[3]

if int(dzien[0])==0 and len(dzien)==2:
    dzien=dzien[1]
    
if int(miesiac[0])==0 and len(miesiac)==2:
    miesiac=miesiac[1]

ostatni_dzien=miesiac+'/'+dzien+'/'+rok
print('Ostatni dzień przed prognozą = ',ostatni_dzien)



i=0
j=0
for i in range(0,len(daty)):
    if ostatni_dzien == daty[i]: #Szukamy indeksu odpowiadającego danej dacie
        j=i
    i=i+1

if j!=0:    #Sprawdzamy, czy istnieje szukana data, w przeciwnym wypadku zwraca komunikat
    #Zapisujemy dane wykorzystane do prognozy
    print('\nPobrany zakres:')
    print('Dzień','Liczba zachorowań')
    dane_do_prognozy=[]
    for i in range(j-int(ile_dni), j):
        dane_do_prognozy.append((daty[i+1],polska[i+1]))
        print((daty[i+1],polska[i+1]))

    #Tworzymy listę z prognozowanymi datami
    print('\nPrognozowany zakres:')
    print('Dzień','Liczba zachorowań')
    dane_prognozowane=[]
    for i in range(int(ile_dni_prognozy)):
        data=datetime.date(int(rok)+2000, int(miesiac), int(dzien)+1)+datetime.timedelta(days = i)
        data=str(data)
        if data[5]==0 :
            data1=data[6]
        else :
            data1=data[5]+data[6]
        if data[8]==0 :
            data2=data[9]
        else :
            data2=data[8]+data[9]
        data3=data[9]
        data3=data[2]+data[3]
        data1=data1.strip('0')
        data=data1+'/'+data2+'/'+data3
        dane_prognozowane.append((data,0))

        print((data,'?'))
else:
    print('Dane z tego dnia nie są jeszcze dostępne')
```

```python
#tworzymy data frame z danymi do prognozy
import pandas as pd
dane_do_prognozy = pd.DataFrame(dane_do_prognozy, columns=["Data", "Liczba zachorowań"])
dane_do_prognozy =dane_do_prognozy.set_index('Data')
print(dane_do_prognozy)
```

```python
#tworzymy data frame z datami które będą prognozowane
dane_prognozowane = pd.DataFrame(dane_prognozowane, columns=["Data", "Liczba zachorowań"])
dane_prognozowane =dane_prognozowane.set_index('Data')
print(dane_prognozowane)
```

```python
#łączymy ramkę danych dane_do_prognozy z ramką danych dane_prognozowane
prognoza=dane_do_prognozy.append(dane_prognozowane)
prognoza
```

```python
weights = [0.05,0.05,0.1,0.1,0.3,0.4]

#prognozujemy poniedziałek i wtorek 
nr_dnia=[0,1,6,7,8,13,14] #nr wierszy z tabeli 'prognoza' potrzebne do przewidywań
p_w=[0,0]
w=[0,0]
k=0
for k in range(0,2):
    p_w[k]=prognoza.iloc[nr_dnia[k:k+6]]["Liczba zachorowań"] #tworzymy dwa wektory danych do prognozowania
    for j in range(0,6):
        w[k]+=p_w[k][j]*weights[j]  #obliczamy średnią kroczącą ważoną
    prognoza.iloc[k+14]['Liczba zachorowań']=round(w[k],0) #uzupełnienie tabeli 'prognoza' o przewidywane dane
    
#prognozujemy środę i czwartek
nr_dnia2=[5,9,10,11,12,13,16]
s_c=[0,0]
w2=[0,0]
k=0
j=0
for k in range(0,2):
    s_c[k]=prognoza.iloc[nr_dnia2[k:k+6]]["Liczba zachorowań"]
    for j in range(0,6):
        w2[k]+=s_c[k][j]*weights[j]
    prognoza.iloc[k+16]['Liczba zachorowań']=round(w2[k],0) 

#prognozujemy piątek, sobotę i niedzielę 
nr_dnia3=[5,6,11,12,13,17,18,19]
p_s_n=[0,0,0]
w3=[0,0,0]
k=0
j=0
for k in range(0,3):
    p_s_n[k]=prognoza.iloc[nr_dnia3[k:k+6]]["Liczba zachorowań"]
    for j in range(0,6):
        w3[k]+=p_s_n[k][j]*weights[j]
    prognoza.iloc[k+18]['Liczba zachorowań']=round(w3[k],0) 
    
#dodanie 20% do danych ze środy, czwartku, piątku i soboty
for k in range(0,4):
    prognoza.iloc[k+16]['Liczba zachorowań']=prognoza.iloc[k+16]['Liczba zachorowań']*1.2 

prognoza.iloc[14]['Liczba zachorowań']=prognoza.iloc[14]['Liczba zachorowań']*0.8 #odjęcie 20% od danych z poniedziałku
    
prognoza[14:21]
```

```python
i=0
j=0
for i in range(0,len(daty)):
    if prognoza.index[20]== daty[i]: #Szukamy indeksu odpowiadającego danej dacie
        j=i
    i=i+1

if j!=0:    #Sprawdzamy, czy istnieje szukana data, w przeciwnym wypadku zwraca komunikat
    #Zapisujemy dane wykorzystane do prognozy
    print('\nPobrany zakres:')
    print('Dzień','Liczba zachorowań')
    rzecz_zach=[]
    daty_prog = []
    for i in range(j-int(ile_dni_prognozy), j):
        rzecz_zach.append(polska[i+1])
        daty_prog.append(daty[i+1])
        print((daty[i+1],polska[i+1]))
        
```

Wyliczamy błąd względny prognozy.

```python
zestawienie =prognoza[14:21]
zestawienie['Rzeczywiste_zach']=rzecz_zach
błąd_wzgl=(zestawienie.loc[:,'Liczba zachorowań']-zestawienie.loc[:,'Rzeczywiste_zach'])/zestawienie.loc[:,'Liczba zachorowań'] * 100
błąd_wzgl
```

* Okres prognozy przypada na czas poświąteczny, zatem dane historyczne, na podstawie których tworzona jest prognoza, są niepełne z uwagi na przypadające święta.
